## This project contains a website with non=tabular data which is converted into tabular data.

#### Below url contains data about various contries and their capital, population and area in km square.

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'http://www.scrapethissite.com/pages/simple/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup.prettify())
```

```python
# get the coulmn names which are under 'strong' tag
table_column_names = soup.find_all('strong')[:3]
```

```python
# clean and convert the headings into a list
table_columns = [column.text.strip() for column in table_column_names]
table_columns
```

```python
# getting the country name and converting the same into list
countries_data = soup.find_all('h3')
country_names = [country.text.strip() for country in countries_data]
country_names
```

```python
# extracting other details that includes capital, population and area.
column_data = soup.find_all('span')
row_data = [data.text.strip() for data in column_data]
each_row = [row_data[i:i+3] for i in range(0, len(row_data), 3)]
each_row
```

```python
# creating a dataframe
df = pd.DataFrame(each_row, columns=table_columns)
```

```python
df.insert(0, 'Country', country_names)
```
